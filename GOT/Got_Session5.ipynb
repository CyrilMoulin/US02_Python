```python
from bs4 import BeautifulSoup
import os
import codecs
import re
```

```python
books_number = {'a game of thrones':1,
                'a clash of kings':2,
                'a storm of swords':3,
                'a feast for crows':4,
                'a dance with dragons':5}

def analyze_page(html):
        soup = BeautifulSoup(html,'html5lib')
        try:
            infobox = soup.find_all('table', class_='infobox')[0]
        except IndexError:
            return
        
        try:
            ibx_name = infobox.caption.get_text().strip()
        except AttributeError:
            ibx_name = None
        
        trs = infobox.find_all('tr')
        for tr in trs:
            try:
                if 'full name' in tr.th.get_text().strip().lower():
                    idx_fullname = tr.td.get_text().strip()
                    
                if 'alias' in tr.th.get_text().strip().lower():
                    aliases = tr.td.get_text().strip()
                    aliases = re.split('\[\d+\]', aliases)
                    regex = r'([a-z])([A-Z])'
                    subst = '\\1|\\2'
                    ibx_aliases = []
                    for alias in aliases :
                        if len(alias.strip()) > 0:
                            alias_clean = re.sub(regex, subst, alias.strip(), 0, re.MULTILINE).split('|')
                            ibx_aliases += alias_clean
                    if len(ibx_aliases) < 1:
                        ibx_aliases = None
                    print(ibx_aliases)
                    
                if 'title' in tr.th.get_text().strip().lower():
                    titles = tr.td.get_text().strip()
                    titles = re.split('\[\d+\]', titles)
                    regex = r'([a-z])([A-Z])'
                    subst = '\\1,\\2'
                    ibx_titles = []
                    for title in titles :
                        if len(title.strip()) > 0:
                            title_clean = re.sub(regex, subst, title.strip(), 0, re.MULTILINE).split(',')
                            ibx_titles += title_clean
                    if len(ibx_titles) < 1:
                        ibx_titles = None
                    print(ibx_titles)
                    
                if 'book' in tr.th.get_text().strip().lower():
                    books = tr.td.get_text().strip()
                    books = re.split('\(\w+\)', books)
                    ibx_books = [books_number[b.strip().lower()] for b in books if len(b.strip())>0 and b.strip().lower() in books_number.keys()]
                    if len(ibx_books) < 1:
                        ibx_books = None
                    print(ibx_books)
                    
            except AttributeError:
                pass
```

```python
for file in os.listdir('awoif/'):
    if file.endswith('.html'):
        # with, garanti que le fichier se ferme à la fin de la boucle
        with open('awoif/' + file, encoding = 'utf-8') as fp:
            html = fp.read()
            analyze_page(html)
```
