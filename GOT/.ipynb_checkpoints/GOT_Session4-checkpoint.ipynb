```python
import codecs
import pickle
import os
import re
from bs4 import BeautifulSoup
```

```python
list_spe = []

for element in os.listdir('awoif/')[1:3]:
    val_char = {}
    html = 'awoif/' + element
    f = codecs.open(html, 'r', 'utf-8')
    soup = BeautifulSoup(f.read())

    links = soup.find_all(dir='auto')
    for link in links:
        name1 = [link.contents[0]]
    val_char['name1'] = name1

    # lstrip() : supprime le premier espace
    # rstrip() : supprime le dernier espace
    links = soup.find_all('caption')
    for link in links:
        name2 = [link.get_text().lstrip().rstrip()]
    val_char['name2'] = name2

    tables = soup.find_all('table', class_='infobox')
    for table in tables:
        trs = table.find_all('tr')
        for tr in trs:
            if tr.th != None:
                values = []
                spe = tr.contents[0].string
                #print(spe)
                if spe == 'Full Name' or spe == 'Alias' or spe == 'Title' or spe == 'Titles':
                    #print(spe)
                    ats = tr.find_all(href=re.compile("index.php"))
                    for a in ats:
                        values.append(a.contents[0].string)
                    if values == []:
                        tds = tr.find_all('td')
                        for td in tds:
                            values.append(td.contents[0].lstrip())
                    val_char[spe] = values
                    #print(val_char)
    list_spe.append(val_char)
print(val_char)
    #print(val_char)
    #print(' ')
print(list_spe)
```

```python
listb = [{'name1': ['Abelar Hightower'], 'name2': ['Abelar Hightower'], 'Title': ['Ser']}, {'name1': ['Addam Frey'], 'name2': ['Addam Frey'], 'Title': ['Ser']}]
with open('list_spe_char.pickle','wb+') as fp:
    pickle.dump(listb, fp)
```
